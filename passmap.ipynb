```python

import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer import Pitch  # Ensure you have mplsoccer installed: pip install mplsoccer

# Create figure and axis
fig, ax = plt.subplots(figsize=(13.5, 8))  # Set figure size




# Set figure and axis background colors
fig.set_facecolor("#223126")  # Background color for the entire figure
ax.set_facecolor("#22312b")  # Background color for the plot area
df =pd.read_csv('Desktop/messibetis.csv')
df['x'] = df['x'] * 1.2  # Multiply 'x' column by 1.2
df['y'] = df['y'] * 0.8  # Multiply 'y' column by 0.8
df['endX'] = df['endX'] * 1.2  # Scale 'endx' by 1.2
df['endY'] = df['endY'] * 0.8  # Scale 'endy' by 0.8

# Create a soccer pitch using mplsoccer's Pitch class
pitch = Pitch(
    pitch_type="statsbomb",  # Corrected pitch type spelling
    
    pitch_color="#22312b",  # Background color of the pitch
    line_color="#c7d5cc",  # Color of the pitch lines
 
   # Ensure proper layout
    
)

# Draw the pitch on the specified axis
pitch.draw(ax=ax)


import matplotlib.pyplot as plt

# Iterate through the DataFrame rows
for x in range(len(df)):  
    # Check if the event was successful
    if df['outcome'][x] == 'Successful':  
        plt.plot(
            (df['x'][x], df['endX'][x]),  # Start and end x-coordinates
            (df['y'][x], df['endY'][x]),  # Start and end y-coordinates
            color="green"
        )
        plt.scatter(df['x'][x], df['y'][x], color="green")  # Mark start point

    # Check if the event was unsuccessful
    elif df['outcome'][x] == 'Unsuccessful':  
        plt.plot(
            (df['x'][x], df['endX'][x]),  
            (df['y'][x], df['endY'][x]),  
            color="red"
        )
        plt.scatter(df['x'][x], df['y'][x], color="red")  # Mark start point

# Add a title
plt.title("messi Pass Map vs Real Betis", color="white", size=14)
ax.scatter([], [], color="green", label="Successful Pass", s=100)
ax.scatter([], [], color="red", label="Unsuccessful Pass", s=100)
ax.legend(loc="upper left", fontsize=12, facecolor="white", edgecolor="white")
 

# Show the plot
plt.show()
# Show the plot
#plt.show()
#plt.gca().invert_yaxis()
```
